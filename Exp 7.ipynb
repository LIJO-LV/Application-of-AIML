```python
import numpy as np
def find_s_algorithm(examples):
    hypothesis = [None] * (len(examples[0]) - 1)  
    for example in examples:
        if example[1] == 'Warm':
            for i in range(len(hypothesis)):
                if hypothesis[i] is None:
                    hypothesis[i] = example[i]
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'
    
    return hypothesis
training_data = [
    ['Sunny', 'Warm', 'Normal', 'Strong','Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong',  'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong',  'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong',  'Change', 'Yes'],
    ['Rainy', 'Warm', 'Normal','Weak', 'Change','Yes']
]
hypothesis = find_s_algorithm(training_data)
print("The most specific hypothesis is:", hypothesis)
```
